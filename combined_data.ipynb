```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fantasy_df = pd.read_csv("goodreads_fan_books_clean.csv")
scifi_df = pd.read_csv("goodreads_sf_books_clean.csv")
```

```python
fantasy_df = fantasy_df.drop(columns=["Unnamed: 0"])
scifi_df = scifi_df.drop(columns=["Unnamed: 0"])
```

```python
fantasy_df["genre"] = "Fantasy"
scifi_df["genre"] = "Sci-Fi"
```

```python
books_df = pd.concat([fantasy_df, scifi_df], ignore_index=True)
```

```python
books_df.head()
```

```python
books_df.info()
```

```python
books_df.describe()
```

```python
books_df.isnull().sum()
```

```python
books_df["series_name"] = books_df["series_name"].fillna("Standalone")
```

```python
books_df["is_series"] = np.where(
    books_df["series_name"] == "Standalone",
    "Standalone",
    "Series"
)
```

```python
books_df.groupby("genre")["avg_rate"].mean()
```

```python
books_df.groupby("genre")["num_rate"].mean()
```

```python
books_df.groupby("is_series")["avg_rate"].mean()
```

```python
books_df.groupby("author")["num_rate"].sum() \
    .sort_values(ascending=False) \
    .head(10)
```

```python
books_df.groupby(["pub_year", "genre"]).size()
```

```python
avg_ratings = books_df.groupby("genre")["avg_rate"].mean()

plt.figure(figsize=(6,5))

plt.bar(avg_ratings.index, avg_ratings.values)

plt.title("Average Goodreads Rating by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Rating")

plt.show()
```

```python
books_df["pub_year"] = pd.to_numeric(
    books_df["pub_year"],
    errors="coerce"
)

filtered_books = books_df[books_df["pub_year"] >= 1940]

yearly_counts = filtered_books.groupby(["pub_year", "genre"]) \
    .size() \
    .unstack()

all_years = range(1940, int(yearly_counts.index.max()) + 1)


yearly_counts.plot(figsize=(12,6))

plt.title("Publication Trends by Genre (1940–Present)")
plt.xlabel("Publication Year")
plt.ylabel("Number of Books Published")

plt.xlim(1940, yearly_counts.index.max())

plt.show()
```

```python
series_ratings = books_df.groupby("is_series")["avg_rate"].mean()

plt.figure(figsize=(6,5))

plt.bar(series_ratings.index, series_ratings.values)

plt.title("Average Ratings: Series vs Standalone")
plt.xlabel("Book Type")
plt.ylabel("Average Rating")

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    books_df["num_rate"],
    books_df["avg_rate"],
    alpha=0.5
)

plt.title("Popularity vs Average Rating")
plt.xlabel("Number of Ratings")
plt.ylabel("Average Rating")

plt.show()
```
